```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Create an empty graph with no nodes and no edges.
G = nx.Graph()
```

```python
# Add a single node
G.add_node(1)
G
```

```python
# Add nodes from a list
G.add_nodes_from([2, 3])
```

```python
# Draw the graph using nx.draw()
plt.figure(figsize=(3,3))
nx.draw(G)
plt.show()
```

```python
# Add an edge
G.add_edge(1,2)
```

```python
plt.figure(figsize=(3,3))
nx.draw(G)
plt.show()
```

```python
e = (2,3)
G.add_edge(*e) # unpack edge tuple*
```

```python
plt.figure(figsize=(2,2))
nx.draw(G, with_labels = True)
plt.show()
```

A node can be any hashable object (except None), and an edge can be associated with any object x using G.add_edge(n1,n2,object=x).

```python
print(G.nodes())
print(G.edges())
```

```python
print(G)
```

Nodes and edges can also contain metadata which can be revealed by passing the 'data' argument as True.

```python
# Create a new node (with attributes 'color' and 'size') and a new edge (with attribute 'weight')
G.add_node(99, color='blue', size='big')
G.add_edge(1, 99, weight=1.1)
print(G.nodes(data=True))
print(G.edges(data=True))
```

```python
G.nodes()[1]
```

```python
# Add attributes to existing nodes
G.nodes()[1]['color'] = 'red'
G.nodes()[2]['color'] = 'blue'
G.nodes()[3]['color'] = 'yellow'
print(G.nodes(data=True))
```

```python
# Finding blue nodes
blue_nodes = [n for n, d in G.nodes(data=True) if d['color'] and d['color'] == 'blue']
blue_nodes
```
